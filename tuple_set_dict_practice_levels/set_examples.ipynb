### LOW LEVEL

```python
s1 = {1, 2, 3}
for i in s1:
    print(i)
```

```python
s2 = {1, 2, 2, 3}
print(s2)
```

```python
s3 = {"apple", "banana"}
s3.add("cherry")
print(s3)
```

### MEDIUM LEVEL

```python
a = {1, 2, 3} # union
b = {3, 4, 5}
print(a | b)
```

```python
print(a & b) # intersection
```

```python
print(a - b) # difference
```

```python
s7 = {10, 15, 20} # loop + condition
for i in s7:
    if i > 12:
        print(i)
```

### HIGH LEVEL

```python
s8 = {1, 2, 3} # remove vs discard
s8.discard(5)
print(s8)
```

```python
a = {1, 2, 3} # symmetric difference
b = {3, 4, 5}
print(a ^ b)
```

```python
s10 = {x*x for x in range(5)}
print(s10)
```
