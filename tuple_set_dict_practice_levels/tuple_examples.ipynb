### LOW LEVEL

```python
t1 = (1, 2, 3)
for i in t1:
    print(i)
```

```python
t2 = ("a", "b", "c")
print(t2[1])
```

```python
t3 = (10, 20, 30, 10)
print(t3.count(10))
```

### MEDIUM LEVEL

```python
t4 = ("Varnika", 22, "MCA") # unpacking
name, age, course = t4
print(name, age)
```

```python
t5 = (10, 20, 30) # loop with index
for i in range(len(t5)):
    print(i, t5[i])
```

```python
t6 = (1, (2, 3), 4) # nested tuple
print(t6[1][0])
```

```python
t7 = (5, 10, 15) # tuple + sum
total = 0
for i in t7:
    total += i
print(total)
```

### HIGH LEVEL

```python
lst = [1, 2, 3] # convert list to tuple
t8 = tuple(lst)
print(t8)
```

```python
t9 = (10, 20, 30, 40) # tuple slicing
print(t9[1:3])
```

```python
t10 = (5, 8, 12, 3) # tuple in loop with condition
for i in t10:
    if i % 2 == 0:
        print(i)
```
