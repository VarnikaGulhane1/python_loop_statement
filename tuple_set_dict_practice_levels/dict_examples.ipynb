### LOW LEVEL

```python
d1 = {"name": "Varnika", "age": 22}
for key in d1:
    print(key)
```

```python
d2 = {"a": 1, "b": 2}
print(d2["a"])
```

```python
d3 = {"x": 10, "y": 20}
print(d3.keys())
```

### MEDIUM LEVEL

```python
d4 = {"a": 1, "b": 2} # items loop
for k, v in d4.items():
    print(k, v)
```

```python
d5 = {"x": 10} # update
d5["y"] = 20
print(d5)
```

```python
d6 = {"a": 1, "b": 2} # pop
d6.pop("a")
print(d6)
```

```python
d7 = {"m": 10, "n": 20} # sum of values
total = 0
for v in d7.values():
    total += v
print(total)
```

### HIGH LEVEL

```python
d8 = {x: x*x for x in range(5)} # dictionary comprehension
print(d8)
```

```python
d9 = {
    "student1": {"name": "A", "age": 20},
    "student2": {"name": "B", "age": 21}
} # nested dictionary
print(d9["student1"]["name"])
```

```python
d10 = {"a": 5, "b": 15, "c": 25} # filter dictionary
for k, v in d10.items():
    if v > 10:
        print(k, v)
```
