# .Keys loop

```python
student = {
    "name": "Sagar",
    "age": 27,
    "course": "Computer Science"
}

for stud in student.keys():
    print(f"stud = {stud} , type of stud = {type(stud)} keyValue = {student.get(stud)}")
```

# .Values for 

```python
student = {
    "name": "Sagar",
    "age": 27,
    "course": "Computer Science"
}

for stud in student.values():
    print(f"stud = {stud} , type of stud = {type(stud)} keyValue = {student.get(stud)}")
```
