### .Keys loop
- to add a level of security (if it will not  the dict it will crash the code at the moment)

```python
student = {
    "name": "Varnika",
    "age": 22,
    "course": "MCA"
}

for key in student.keys():
    print(f"key = {key}, type of key = {type(key)}, keyValue = {student.get(key)}")
```

### .Values for

```python
student = {
    "name": "Varnika",
    "age": 22,
    "course": "MCA"
}

for key in student.values():
    print(f"key = {key}, type of key = {type(key)}, keyValue = {student.get(key)}")
```
