Here’s a clear and simple example of using a **for loop with a Python dictionary** 👇

### 🔹 Basic Example

```python
student = {
    "name": "Varnika",
    "age": 22,
    "course": "MCA"
}

# Loop through keys
for key in student:
    print(key)
```

👉 Output:

```
name
age
course
```

---

### 🔹 Loop through values

```python
for value in student.values():
    print(value)
```

👉 Output:

```
Varnika
22
MCA
```

---

### 🔹 Loop through key-value pairs (Most Important ⭐)

```python
for key, value in student.items():
    print(key, ":", value)
```

👉 Output:

```
name : Varnika
age : 22
course : MCA
```

---

### 🔹 With condition

```python
for key, value in student.items():
    if key == "age":
        print("Age found:", value)
```

---

### 🔹 Real-life style example

```python
marks = {
    "Math": 90,
    "Science": 85,
    "English": 88
}

for subject, score in marks.items():
    print(f"{subject} -> {score}")
```
