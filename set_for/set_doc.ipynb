# 📘 Set in Python – Documentation

## 🔹 Definition

A **set** is a built-in Python data type used to store multiple items in a single variable.
Sets are **unordered, mutable, and do not allow duplicate values**.

---

## 🔹 Key Features

* ❌ Unordered (no indexing)
* ❌ No duplicate elements
* ✅ Mutable (can be modified)
* ✅ Supports mathematical operations (union, intersection, etc.)

---

## 🔹 Syntax

```python
set_name = {element1, element2, element3}
```

---

## 🔹 Example

```python
fruits = {"apple", "banana", "cherry"}
print(fruits)
```

👉 Output (order may vary):

```
{'banana', 'apple', 'cherry'}
```

---

## 🔹 Creating a Set

```python
# Using curly braces
s1 = {1, 2, 3}

# Using set() function
s2 = set([1, 2, 3])
```

---

## 🔹 Duplicate Values (Important ⭐)

```python
numbers = {1, 2, 2, 3, 3}
print(numbers)
```

👉 Output:

```
{1, 2, 3}
```

---

## 🔹 Accessing Elements

❌ Sets do NOT support indexing
✅ Use loop instead:

```python
s = {"a", "b", "c"}

for item in s:
    print(item)
```

---

## 🔹 Adding Elements

```python
s = {1, 2, 3}

s.add(4)
print(s)
```

---

## 🔹 Removing Elements

```python
s = {1, 2, 3}

s.remove(2)   # Error if not found
s.discard(5)  # No error if not found
print(s)
```

---

## 🔹 Set Operations (Very Important ⭐)

### 1. Union

```python
a = {1, 2, 3}
b = {3, 4, 5}

print(a | b)
```

👉 Output:

```
{1, 2, 3, 4, 5}
```

---

### 2. Intersection

```python
print(a & b)
```

👉 Output:

```
{3}
```

---

### 3. Difference

```python
print(a - b)
```

👉 Output:

```
{1, 2}
```

---

### 4. Symmetric Difference

```python
print(a ^ b)
```

👉 Output:

```
{1, 2, 4, 5}
```

---

## 🔹 Set Methods

```python
s = {1, 2, 3}

s.add(4)        # Add element
s.remove(2)     # Remove element
s.pop()         # Remove random element
s.clear()       # Remove all elements
```

---

## 🔹 Frozen Set (Immutable Set)

```python
fs = frozenset([1, 2, 3])
# fs.add(4) ❌ Error (cannot modify)
```

---

## 🔹 Real-Life Example

```python
students = {"Amit", "Varnika", "Riya"}
new_students = {"Riya", "Karan"}

# Find all unique students
all_students = students | new_students
print(all_students)
```

---

## 🔹 Advantages of Set

* Removes duplicates automatically
* Fast membership testing
* Useful for mathematical operations

---

## 🔹 Difference: Set vs List

| Feature    | Set           | List      |
| ---------- | ------------- | --------- |
| Order      | ❌ Unordered   | ✅ Ordered |
| Duplicates | ❌ Not allowed | ✅ Allowed |
| Indexing   | ❌ No          | ✅ Yes     |
| Mutable    | ✅ Yes         | ✅ Yes     |

---

## 🔹 Conclusion

Sets are useful when:

* You need **unique values only**
* Performing **mathematical operations**
* Checking membership quickly
