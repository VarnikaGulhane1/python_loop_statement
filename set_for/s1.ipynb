```python
s1 = {1, 2, 3}
s2 = {1, 2, 3, 5}

print(s1 | s2)
```

```python
s1 = {1, 2, 3, 3, 8}
s2 = {1, 2, 3, 5, 8}

print(f"union = {s1 | s2}")
print(f"intersect = {s1 & s2}")
print(f"Difference = {s1 - s2}")
print(f"Difference = {s2 - s1}")
```
