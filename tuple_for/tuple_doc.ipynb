# 📘 Tuple in Python – Documentation

## 🔹 Definition

A **tuple** is a built-in Python data type used to store multiple items in a single variable.
Tuples are **ordered, immutable, and allow duplicate values**.

---

## 🔹 Key Features

* ✅ Ordered (maintains insertion order)
* ❌ Immutable (cannot be changed after creation)
* ✅ Allows duplicate elements
* ✅ Can store multiple data types

---

## 🔹 Syntax

```python
tuple_name = (element1, element2, element3)
```

---

## 🔹 Example

```python
student = ("Varnika", 22, "MCA")
print(student)
```

👉 Output:

```
('Varnika', 22, 'MCA')
```

---

## 🔹 Accessing Elements

```python
student = ("Varnika", 22, "MCA")

print(student[0])   # First element
print(student[-1])  # Last element
```

👉 Output:

```
Varnika
MCA
```

---

## 🔹 Tuple with Different Data Types

```python
data = ("Varnika", 22, 85.5, True)
print(data)
```

---

## 🔹 Single Element Tuple

```python
single = (10,)
print(type(single))
```

👉 Important: Without comma `(10)` is not a tuple.

---

## 🔹 Looping Through a Tuple

```python
numbers = (1, 2, 3, 4)

for num in numbers:
    print(num)
```

---

## 🔹 Tuple Packing and Unpacking

```python
# Packing
person = ("Varnika", 22, "MCA")

# Unpacking
name, age, course = person

print(name)
print(age)
print(course)
```

---

## 🔹 Tuple Methods

Tuples have only two built-in methods:

### 1. count()

```python
t = (1, 2, 2, 3)
print(t.count(2))
```

👉 Output: `2`

### 2. index()

```python
t = (1, 2, 3)
print(t.index(2))
```

👉 Output: `1`

---

## 🔹 Immutability (Important ⭐)

```python
t = (1, 2, 3)
t[0] = 10   # ❌ This will give error
```

---

## 🔹 Nested Tuple

```python
nested = (1, (2, 3), (4, 5))
print(nested[1])
```

---

## 🔹 Advantages of Tuple

* Faster than lists
* Safe (data cannot be modified)
* Can be used as dictionary keys

---

## 🔹 Difference: Tuple vs List

| Feature | Tuple  | List   |
| ------- | ------ | ------ |
| Mutable | ❌ No   | ✅ Yes  |
| Syntax  | `( )`  | `[ ]`  |
| Speed   | Faster | Slower |
| Methods | Few    | Many   |

---

## 🔹 Real-Life Example

```python
coordinates = (19.0760, 72.8777)  # Latitude, Longitude
print("Location:", coordinates)
```

---

## 🔹 Conclusion

Tuples are useful when:

* Data should not be changed
* You want faster performance
* Fixed collections (like coordinates, days, etc.)
