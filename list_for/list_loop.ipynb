### Correct logic

```python
l2 = list(range(1, 100))

prvValue = 0
for l2Value in l2:
    currAdd = prvValue + l2Value
    prvValue = currAdd
    print(f"prvValue = {prvValue}, currAdd = {currAdd}, l2Value = {l2Value}")
    print()
```
